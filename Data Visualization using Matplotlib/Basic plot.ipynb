```python
import matplotlib.pyplot as plt
```

US population over years.

```python
Year=[2014,2015,2016]
Population=[319,322,324]
```

```python
plt.plot(Year,Population)
```

```python
plt.show()
```

Now let's add labels

```python
plt.plot(Year,Population)
Year=[2014,2015,2016]
Population=[319,322,324]
plt.xlabel('Year')
plt.ylabel('Population (Million)')
plt.show()
```

add title

```python
plt.plot(Year,Population)
Year=[2014,2015,2016]
Population=[319,322,324]
plt.xlabel('Year')
plt.ylabel('Population (Million)')
plt.title("Us population over years")
plt.show()
```

Add ticks in the x-axis

```python
plt.plot(Year,Population)
Year=[2014,2015,2016]
Population=[319,322,324]
plt.xlabel('Year')
plt.ylabel('Population (Million)')
plt.title("Us population over years")
plt.xticks([2014, 2015, 2016])
plt.show()
```
