## Introduction

The purpose of a plotting package is to help us build visual aids for our data as easily as possible, with all the necessary control – that is, by focussing on the larger picture (requirements) i.e using relatively high-level commands most of the time and not on the minute details, and still have the ability to use the low-level commands when needed.

matplotlib is a very vast codebase that could be taken advantage of, if there is a clear understanding of the conceptual framework and knowledge of a few important points.

Everything in matplotlib is organized in a hierarchy. 

1. At the top of the hierarchy is the matplotlib “state-machine environment” which is provided by the matplotlib.pyplot module. 

2. At this level, simple functions are used to add plot elements (lines, images, text, etc.) to the current axes in the current figure.

3. The next level down in the hierarchy is the first level of the object-oriented interface, in which pyplot is used only for a few functions such as figure creation, and the user explicitly creates and keeps track of the figure and axes objects. 

4. At this level, the user uses pyplot to create figures, and through those figures, one or more axes objects can be created. These axes objects are then used for most plotting actions.



matplotlib.pyplot is a collection of command style functions that make matplotlib work like MATLAB. 

The various states such as current figure or plotting area are kept track of in matplotlib.pyplot and the plotting functions are directed to the current axes.

The current axes is not referring to the mathematical "plural" of axis, but instead the plot itself. It is the region of the image with the data space. The Axes contains two (or three in the case of 3D) Axis objects (be aware of the difference between Axes and Axis) which take care of the data limits.

#### Lets jump right into our first example

```python
import matplotlib.pyplot as plt
plt.plot([1,2,3,4]) # the dat to be plotted
plt.ylabel('y axis') # names for the axes.
plt.xlabel('x axis')
plt.show() # Finally show() will print the plot on screen
```

The curios thing to notice here is that the x-axis ranges from 0-3 and the y-axis from 1-4. But we never provided any data for the x axis.

How then?

Turns out that if we provide a single list or array to the plot() command, matplotlib assumes it is a sequence of y values, and generates the corresponding x values automatically.

Since python ranges start with 0, the default x vector has the same length as y but starts with 0. Hence the x data are [0,1,2,3].

______________

plot() is a versatile command, and will take an arbitrary number of arguments. For example, to plot x versus y, you can issue the command:

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
```

```python
# We need to say show to see the plot
plt.show()
```

For every x, y pair of arguments, there is an optional third argument which is the format string that indicates the color and line type of the plot.

```python
plt.plot([1,2,3,4], [1,4,9,16], 'ro') # r represents red and o represents circles
plt.axis([0, 6, 0, 20])
plt.show()
```

The format string is adopted from MATLAB where we concatenate a color string with a line style string. The default format string is ‘b-‘, which is a solid blue line. 

For example, to plot the above with green circles, we would say:

```python
plt.plot([1,2,3,4], [1,4,9,16], 'go') # r represents red and o represents circles
plt.axis([0, 6, 0, 20])
plt.show()
```

### Use of Numpy
In general, in Python numpy is used to handle data in array like fashion. In fact, all sequences are converted to numpy arrays internally.

```python
import numpy as np
# evenly sampled time at 200ms intervals
t = np.arange(0, 2, 0.1)

# red dashes, blue squares and green triangles
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```

### Controlling line properties
Lines have many attributes that you can set: linewidth, dash style, antialiased, etc. There are several ways to set line properties.

```python
# Use keyword args:
x = np.array([1,2,3,4])
y = np.array([4,16,64,256])
plt.plot(x, y, linewidth=4.0)
plt.show()
```

##### Use the setp() command. 

We'll see how to use MATLAB-style command to set multiple properties on a list of lines. 

setp() works transparently with a list of objects or a single object. You can either use python keyword arguments or MATLAB-style string/value pairs:

```python
lines = plt.plot(x, y)
# use keyword args
plt.setp(lines, color='r', linewidth=2.0)
plt.show()
```

```python
lines = plt.plot(x, y)
# or MATLAB style string value pairs
plt.setp(lines, 'color', 'g', 'linewidth', 5.0)
plt.show()
```

### Working with multiple figures and axes

```python
# the function returns a mathematical function
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure(1)
plt.subplot(211)
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')

plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.show()
```
